# PCA 

```python
# We will Apply PCA analysis to reduce the dimentiality of the data
```

```python
from sklearn.decomposition import PCA
```

```python
# Number of observations
n = features.shape[0]
# Number of features
p = features.shape[1]
n,p
```

```python
# Run the PCA 
pca = PCA()
pca = pca.fit(features)
```

```python
coord = pca.fit_transform(normalized_data)
#number of calculated components
print(pca.n_components_) 
```

```python
#adjusted values
eigval = (n-1)/n*pca.explained_variance_
```

```python
#scree plot of Eigen values % Factor number
plt.plot(np.arange(1,p+1),eigval)
plt.title("Scree plot")
plt.ylabel("Eigen values")
plt.xlabel("Factor number")
plt.show()
```

```python
#cumulative explained variance
variance = pd.DataFrame(pca.explained_variance_ratio_)
np.cumsum(pca.explained_variance_ratio_)
```

```python
#Plot the cumulative explained variance
plt.plot(np.arange(1,p+1),np.cumsum(pca.explained_variance_ratio_))
plt.title("Explained variance vs. # of factors")
plt.ylabel("Cumsum explained variance ratio")
plt.xlabel("Factor number")
plt.show()
```

```python
# Seuils to define the number of componenets (test des bÃ¢tons brisÃ©s)
bs = 1/np.arange(p,0,-1)
bs = np.cumsum(bs)
bs = bs[::-1]
```

```python
print(pd.DataFrame({'Val.Propre':eigval,'Seuils':bs}))
```

```python
# Based on the cumul variance we will choose 8 axes, with 6 components we have total varaince of 84%
```

```python
# Apply PCA with 6 components 
pca = PCA(6)
pca = pca.fit(features)
dataPCA = pca.transform(features)
```

```python
datafPCA=pd.DataFrame(dataPCA)
```

# Discretization 

```python
from sklearn.preprocessing import KBinsDiscretizer
```

```python
# histogram of the raw data
Value=dataf.Value
Val=pd.DataFrame({"Value":Value})
plt.hist(Val,bins=5)
plt.show()
```

```python
Val=pd.DataFrame({"Value":dataf.Value}).to_numpy()

# reshape data to have rows and columns
Val = Val.reshape((len(Val),1))
# discretization transform the raw data into 5 categories 
kbins = KBinsDiscretizer(n_bins=5, encode='ordinal', strategy='uniform')
data_trans = kbins.fit_transform(Val)
```
